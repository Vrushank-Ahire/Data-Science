### K Nearest Neighbour Classfier

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
X , y = make_classification(n_samples=1000,n_features=3,
                            n_redundant=1,
                            n_classes=2,random_state=999)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    X,y,random_state=42,test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import (accuracy_score,classification_report,confusion_matrix)
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
###Task use grid search CV here and also use for loop
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
k_params = []

for i in range(10):
    k_params.append(i+1)
    
def foo(X_train,X_test,y_train,y_test,neigh_num):
    
    report = {}
    
    for i in range(neigh_num):
        ##Create classfier
        classifier = KNeighborsClassifier(n_neighbors= (i+1))
        classifier.fit(X_train,y_train)
        y_pred = classifier.predict(X_test)
        score = accuracy_score(y_test,y_pred)
        report[k_params[i]] = score
    
    return report 
```

```python
report = foo(X_train,X_test,y_train,y_test,10)
```

```python
values_accu = list(report.values())
```

```python
max(values_accu)
```

```python
##Grid search CV
```

```python
parameters = {'n_neighbors' : k_params ,
             'weights' : ['uniform','distance'] ,
             'p' : [1,2] ,
            'algorithm' : ['ball_tree','kd_tree','auto'],
              'metric': ['minkowski', 'manhattan', 'euclidean']
            }

from sklearn.model_selection import GridSearchCV
clf  = GridSearchCV(KNeighborsClassifier(),param_grid =parameters ,cv =3,scoring = 'accuracy',verbose=3)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
