```python
#Independent and dependent features

X = df.iloc[ : , : -1]
y = df.iloc[ : , -1]
```

```python
X.head()
```

```python
y.value_counts()
```

```python
## train test split 
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

#####  Note there is no need to perform standardization in decision tree becasuse the small values will not put any impact on the tree complexity , in short no need to perform feature scaling

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(treeclassifier,max_depth=2,filled=True)
```

```python
###post Pruning with max depth = 2
treeclassifier = DecisionTreeClassifier(criterion='entropy',max_depth=2)
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(5,5))
tree.plot_tree(treeclassifier,filled=True)
```

```python
##prediction
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
print(classification_report(y_test,y_pred))
```

```python
### Internal Assignment
from sklearn.datasets import load_breast_cancer
```

```python
dataset2 = load_breast_cancer()
```

```python
print(dataset2.DESCR)
```

```python
column_names = list(dataset2.feature_names)
len(column_names)
```

```python
type(dataset2.data)
```

```python
DF = pd.DataFrame(dataset2.data,columns=column_names)
```

```python
len(DF.count())
```

```python
DF['target'] = dataset2.target
```

```python
len(list(DF.columns))
```

```python
DF.target.value_counts()
```

```python
DF.head()
```

```python
##creating input and output features
X = DF.iloc[: , :-1]
y = DF.iloc[: ,-1]
```

```python
X.shape,y.shape
```

```python
##train test split
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test ,y_train ,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
TreeClassifier = DecisionTreeClassifier()
```

```python
TreeClassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(TreeClassifier,filled=True)
```

```python
y_pred1 = TreeClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_pred1,y_test))
print(classification_report(y_pred1,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
TreeClassifier2 = DecisionTreeClassifier(max_depth=4)   #max depth 3 gives more accuracy than this
```

```python
TreeClassifier2.fit(X_train,y_train)
```

```python
plt.figure(figsize = (15,10))
tree.plot_tree(TreeClassifier2,filled=True)
```

```python
y_pred2 = TreeClassifier2.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

#### DECISION Tree Prepruning and Hyperparameter tuning For huge data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth': [1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
clf = GridSearchCV(classifier,param_grid=parameters,cv=5,scoring='accuracy')
```

```python
clf
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
