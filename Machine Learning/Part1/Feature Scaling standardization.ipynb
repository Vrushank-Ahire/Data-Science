```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill = list(df['total_bill'])
```

```python
total_bill
```

```python
sns.histplot( total_bill)
```

formula Z_score = (Xi - X_(mean vlaue)) / std

```python
import numpy as np
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean,std
```

```python
normalised_data = []
for i in total_bill:
    z_score = (i - mean)/std
    normalised_data.append(z_score)
```

```python
normalised_data
```

```python
sns.histplot(normalised_data)
```

###### the above manual process involves for loops which would be costly for a million size list
###### so its better to use inbuilt libraries 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler   #object of StandardScalar class
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
scaler.get_feature_names_out()
```

```python
scaler.transform(df[['total_bill','tip']])
```

```python
import pandas as pd
df = pd.DataFrame(scaler.transform(df[['total_bill','tip']]))
```

```python
df.rename(columns={0 :'total_bill',1 : 'tips'},inplace=True)
```
