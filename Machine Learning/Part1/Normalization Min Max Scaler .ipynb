```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

### basic formula used for normalization is 
    
    (Xi - Xmin) / (Xmax - Xmin)

```python
import numpy as np
```

```python
lst_fare = list(df['fare'])
```

```python
min_fare = np.min(lst_fare)
max_fare = np.max(lst_fare)

normalised_fare_lst = []
for i in lst_fare:
    normalised_val = (i - min_fare)/(max_fare - min_fare)
    normalised_fare_lst.append(normalised_val)
```

```python
##normalised_fare_lst
```

```python
sns.scatterplot(normalised_fare_lst)
```

```python
sns.histplot(normalised_fare_lst)
```

##### THIS PROCESS canbe simply performed using a 
##### library sklearn .preprocessing class MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max
```

```python
min_max.fit(df[['fare','tip','distance']])
```

```python
min_max.transform(df[['fare','tip','distance']])
```

```python
##directly fit transform can be also used instead of fit and transform indivisually
min_max.fit_transform(df[['fare','tip','distance']])
```

```python
##if calling for new value set no need to fit again , directly use transform
min_max.transform([[7,2,5]]) 
```
